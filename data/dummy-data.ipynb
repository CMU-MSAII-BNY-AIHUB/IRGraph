```python
import xml.etree.ElementTree as ET
import json

xml_file_path = "The Bank of New York Mellon Corporation, Q2 2023 Earnings Call, Jul 18, 2023 (1).xml"
tree = ET.parse(xml_file_path)


# Parse the XML
root = tree.getroot()

# Extract title from header


# Extract person tags
named_entities = []
for person in root.findall('.//person'):
    title = person.get('position') if person.get('position') is not None else ""
    entity = {
        "value": person.text,
        "label": "person",
        "attribute": {"title": title}
    }
    named_entities.append(entity)

# Convert to JSON
json_data = json.dumps(named_entities, indent=4)

# Show the JSON data (for demonstration)
print(json_data)

# Save to JSON file
json_filename = 'dummy_data.json'
with open(json_filename, 'w') as file:
    file.write(json_data)

json_filename

```
